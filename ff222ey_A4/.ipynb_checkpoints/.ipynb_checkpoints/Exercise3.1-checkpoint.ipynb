```python
from ipynb.fs.full.Exercise2 import sammon
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn import preprocessing
from IPython.display import HTML
```

```python
from sklearn.datasets import load_digits
digits = load_digits()

from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()

from sklearn.datasets import load_wine
wine = load_wine()
```

```python
from sklearn.decomposition import PCA

from sklearn.manifold import TSNE
```

```python
datasets = [data, wine, digits]
drs = [PCA, TSNE, sammon]
datasets_name = ['Breast cancer', 'Wine', 'Mnist 0-9']
drs_name = ['PCA', 'TSNE', 'Sammon']
```

```python
itera = 2000
thresh = 1e-9
alpha = 0.3
```

```python
font = {'family' : 'normal',
        'size'   : 14}

plt.rc('font', **font)

if hasattr(TSNE, 'fit_transform'):
    print('True')

fig1, ax1 = plt.subplots(nrows=3, ncols=3)
i = 0
for row in ax1:
    data = datasets[i]
    X = preprocessing.normalize(data.data, axis = 0)
    y = data.target
    X, y = X[:300], y[:300]
    j = 0
    for col in row:
        dr = drs[j]
        if hasattr(dr, 'fit_transform'):
            Xe = dr(n_components= 2).fit_transform(X)
        else:
            Xe = dr(X,itera, thresh, alpha, 1e-4)
        col.scatter(Xe[:,0], Xe[:,1], alpha=0.8, c=y, s=16, cmap='viridis')
        if j == 0:
            col.set_ylabel(datasets_name[i])
        if i == 0:
            col.set_title(drs_name[j])
        col.set_xticks([]),col.set_yticks([])
        j +=1
    i += 1
plt.tight_layout()
fig1.set_size_inches(13, 15)
```

# Comparison of DR Techniques

The datasets used for comprarisons of the different Dr techniques were the Breast cancer, Wine and MNiST dataset. The Breast cancer dataset contained 569 observations, but only 300 was used, it also contained 30 features with 2 different classes. The Wine dataset contained 178 observations and 13 features with 3 classes. The MNIST dataset was a bolied down version with only 64 features. The original dataset contained 1797 observations but once again only 300 observations were used with 10 different classes (0-9).

When using the TSNE dataset it is hard to call a clear winner, All 3 dr techniques performed well when seperating the classes and the same goes for the Wine dataset. However when the MNIST dataset was used we can cleary see that TSNE was by far the dr technique that perfomed the best.
