```python
import numpy as np
from sklearn import preprocessing
from sklearn.manifold import TSNE

from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
import import_ipynb
from Exercise1 import bkmeans
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_digits
digits = load_digits()

from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()

from sklearn.datasets import load_wine
wine = load_wine()
```

```python
datasets = [data, wine, digits]
clusterTechniques = [KMeans, bkmeans, AgglomerativeClustering]
clusterNames = ['Kmeans', 'bkmeans', 'Agglomerative']
datasets_name = ['Breast cancer', 'Wine', 'Mnist 0-9']
```

```python
fig1, ax1 = plt.subplots(nrows=3, ncols=3)
i = 0
for row in ax1:
    data = datasets[i]
    clusters = len(data.target_names)
    X = preprocessing.normalize(data.data, axis = 0)
    y = data.target
    Xe = TSNE(n_components= 2).fit_transform(X)
    j = 0
    for col in row:
       cluster = clusterTechniques[j]
       if hasattr(cluster, 'fit_predict'):
           y = cluster(n_clusters = clusters).fit_predict(Xe)
       else:
           y = cluster(Xe, k=clusters, itera = 10)
       col.scatter(Xe[:,0], Xe[:,1], alpha=0.8, c=y, s=16, cmap='viridis')
       if j == 0:
           col.set_ylabel(datasets_name[i])
       if i == 0:
           col.set_title(clusterNames[j])
       col.set_xticks([]),col.set_yticks([])
       j +=1
    i += 1
              
plt.tight_layout()
fig1.set_size_inches(15, 12)
```

# Comparison of Cluster Techniques.

For this exercise, TSNE was first used on all the datasets. On the Breast cancer dataset, all 3 clustering techniques pretty much gave the same results. On the wine dataset, it is hard to see a good seperation since all the classes are pretty much close to each other in a straight line, But i would say that KMeans did a slighly better job on evenly distributing the classes. Moving on to the Mnist dataset, it is also hard to call a clear winner, but i would say that the seperation of the Agglomerative makes a bit more sense.
